```python
 
import pandas as pd 
A = pd.Series([10, 20, 30, 40, 50]) 
B= pd.Series([40, 50, 60, 70, 80]) 
print("Series A:",A) 
print("\nSeries B:",B) 
print("\nSum of Series B:\n", B.sum()) 
print("\nAverage of Series A:\n",A.mean()) 
print("\nMedian of Series B:\n", A.median())
non_com = A[~A.isin(B)].tolist() + B[~B.isin(A)].tolist() 
print("Items not common to both Series:",non_com) 
print("\nSmallest element in Series A:\n", A.min()) 
print("\nLargest element in Series A:\n",A.max()) 
```

```python
import pandas as pd
import matplotlib.pyplot as plt
c= pd.read_csv("titanic.csv")
d = pd.DataFrame(c)
labels = ['Survived', 'Not Survived']
colors = ['r','b']

survived_count = d['Survived'].sum()
not_survived_count = len(d) - survived_count
sizes = [survived_count, not_survived_count]

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.0f%%', startangle=90)
```

```python
import pandas as p 
import matplotlib.pyplot as m 
d=p.read_csv("titanic.csv") 
 
c=d["Pclass"].value_counts() 
co=['g','r','k'] 
m.bar(c.index,c.values,color=co,width=0.5) 
 
m.xticks([1,2,3],["1st class","2nd class","3rd class"]) 
m.xlabel("Classes");
m.ylabel("No of Passengers");
m.title("No of Passengers Travelled in Specific Classes") 
m.show()
```
